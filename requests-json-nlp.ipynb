3. Write a function that takes an artist, song, and filename, accesses the lyrics.ovh api to get the song lyrics, and writes the results to the specified filename.  Test this function by getting the lyrics to any four songs of your choice and storing them in different files.

```python
# Define a function to perform sentiment analysis on song lyrics stored in a file and return the polarity score
def analyze_lyrics_sentiment(filename):
    """
    Perform sentiment analysis on song lyrics stored in a file and return the polarity score.

    Parameters:
        filename (str): The name of the file containing song lyrics.

    Returns:
        float: The polarity score indicating the sentiment of the lyrics.
    """
    # Load spaCy model with TextBlob capabilities
    nlp = spacy.load("en_core_web_sm")
    nlp.add_pipe("spacytextblob")

    # Read the contents of the file
    with open(filename, 'r') as file:
        lyrics = file.read()

    # Perform sentiment analysis on the lyrics
    doc = nlp(lyrics)

    # Get the polarity score of the sentiment analysis
    polarity_score = doc._.polarity

    return polarity_score

# Test the function with the pop songs created in question 3
filename_list = [
    "shape_of_you_lyrics.txt",
    "shake_it_off_lyrics.txt",
    "hello_lyrics.txt",
    "just_the_way_you_are_lyrics.txt"
]

for filename in filename_list:
    song_name = filename.split("_")[0].title()  # Extract the song name from the filename
    polarity_score = analyze_lyrics_sentiment(filename)
    truncated_score = round(polarity_score, 2)  # Truncate the polarity score to two decimal places
    print(f"Polarity score for '{song_name}': {truncated_score}")

```

4. Write a function that takes the name of a file that contains song lyrics, loads the file, performs sentiment analysis, and returns the polarity score.
```python
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

def analyze_lyrics_sentiment(filename):
    """
    Perform sentiment analysis on song lyrics stored in a file and return the polarity score.

    Parameters:
        filename (str): The name of the file containing song lyrics.

    Returns:
        float: The polarity score indicating the sentiment of the lyrics.
    """
    # Load spaCy model with TextBlob capabilities
    nlp = spacy.load("en_core_web_sm")
    nlp.add_pipe("spacytextblob")

    # Read the contents of the file
    with open(filename, 'r') as file:
        lyrics = file.read()

    # Perform sentiment analysis on the lyrics
    doc = nlp(lyrics)

    # Get the polarity score of the sentiment analysis
    polarity_score = doc._.polarity

    # Truncate the polarity score to two decimal places
    truncated_score = round(polarity_score, 2)

    return truncated_score

# Test the function with the three files created in question 3
filename_list = [
    "shape_of_you_lyrics.txt",
    "shake_it_off_lyrics.txt",
    "hello_lyrics.txt",
    "just_the_way_you_are_lyrics.txt"
]

for filename in filename_list:
    song_name = filename.split("_")[0].title()  # Extract the song name from the filename
    polarity_score = analyze_lyrics_sentiment(filename)
    print(f"Polarity score for '{song_name}': {polarity_score}")
```
